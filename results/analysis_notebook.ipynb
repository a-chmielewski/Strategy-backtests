```python
import pandas as pd
```

```python
results = pd.read_csv('merged_results.csv')
results.head()
```

```python
# Find strategies that had 0 trades across all symbol and timeframe combinations
strategies_with_zero_trades = results.groupby('strategy_name')['num_trades'].max()
zero_trade_strategies = strategies_with_zero_trades[strategies_with_zero_trades == 0].index.tolist()

print("Strategies with 0 trades in every symbol and timeframe combination:")
for strategy in zero_trade_strategies:
    print(f"- {strategy}")
    
if not zero_trade_strategies:
    print("No strategies found with 0 trades in all combinations.")
else:
    print(f"\nTotal strategies with 0 trades: {len(zero_trade_strategies)}")
```

```python
# Find top 3 strategy-symbol-timeframe combinations with highest win rate
# Filter out rows with 0 or NaN win rates
filtered_results = results[results['win_rate'] > 0].copy()
filtered_results = filtered_results[filtered_results['win_rate'] < 100].copy()

if len(filtered_results) > 0:
    # Sort by win rate in descending order and get top 3
    top_combinations = filtered_results.nlargest(3, 'win_rate')[['strategy_name', 'coinpair', 'timeframe', 'win_rate', 'leverage', 'num_trades', 'equity_final']]
    
    print("Top 3 Strategy-Symbol-Timeframe combinations with highest win rate:")
    print("=" * 70)
    
    for i, (idx, row) in enumerate(top_combinations.iterrows(), 1):
        print(f"{i}. Strategy: {row['strategy_name']}")
        print(f"   Symbol: {row['coinpair']}")
        print(f"   Timeframe: {row['timeframe']}")
        print(f"   Leverage: {row['leverage']}")
        print(f"   Win Rate: {row['win_rate']:.2f}%")
        print(f"   Total Trades: {int(row['num_trades'])}")
        print(f"   Final Equity: {row['equity_final']:.2f}")
        print("-" * 50)
else:
    print("No combinations found with win rate > 0")
```

```python
# Find top 3 strategy-symbol-timeframe combinations with highest final equity
# Filter out rows with NaN final equity
filtered_results = results[results['equity_final'].notna()].copy()

if len(filtered_results) > 0:
    # Sort by final equity in descending order and get top 3
    top_combinations = filtered_results.nlargest(3, 'equity_final')[['strategy_name', 'coinpair', 'timeframe', 'equity_final', 'win_rate', 'leverage', 'num_trades', 'max_drawdown']]
    
    print("Top 3 Strategy-Symbol-Timeframe combinations with highest final equity:")
    print("=" * 70)
    
    for i, (idx, row) in enumerate(top_combinations.iterrows(), 1):
        print(f"{i}. Strategy: {row['strategy_name']}")
        print(f"   Symbol: {row['coinpair']}")
        print(f"   Timeframe: {row['timeframe']}")
        print(f"   Leverage: {row['leverage']}")
        print(f"   Final Equity: {row['equity_final']:.2f}")
        print(f"   Win Rate: {row['win_rate']:.2f}%")
        print(f"   Total Trades: {int(row['num_trades'])}")
        print(f"   Max Drawdown: {row['max_drawdown']:.2f}%")
        print("-" * 50)
else:
    print("No combinations found with valid final equity values")
```

```python
# Find top 3 strategy-symbol-timeframe combinations with specific criteria
# Filter: win rate > 55%, final equity > 110, sort by most trades
filtered_results = results[
    (results['win_rate'] > 55) & 
    (results['equity_final'] > 110) & 
    (results['num_trades'].notna())
].copy()

if len(filtered_results) > 0:
    # Sort by total trades in descending order and get top 3
    top_combinations = filtered_results.nlargest(3, 'num_trades')[['strategy_name', 'coinpair', 'timeframe', 'num_trades', 'win_rate', 'leverage', 'equity_final', 'max_drawdown']]
    
    print("Top 3 Strategy-Symbol-Timeframe combinations with >55% win rate, >110 final equity, most trades:")
    print("=" * 85)
    
    for i, (idx, row) in enumerate(top_combinations.iterrows(), 1):
        print(f"{i}. Strategy: {row['strategy_name']}")
        print(f"   Symbol: {row['coinpair']}")
        print(f"   Timeframe: {row['timeframe']}")
        print(f"   Total Trades: {int(row['num_trades'])}")
        print(f"   Leverage: {row['leverage']}")
        print(f"   Win Rate: {row['win_rate']:.2f}%")
        print(f"   Final Equity: {row['equity_final']:.2f}")
        print(f"   Max Drawdown: {row['max_drawdown']:.2f}%")
        print("-" * 60)
        
    print(f"\nTotal combinations meeting criteria: {len(filtered_results)}")
else:
    print("No combinations found meeting the criteria:")
    print("- Win rate > 55%")
    print("- Final equity > 110")
    print("- Valid trade data")
```
